```python
import pandas as pd
```

```python
d_df = pd.read_csv("data.csv")
d_df
```

```python
d_df.duplicated().any()
```

```python
d_df = d_df.drop_duplicates()
```

```python
d_df
```

```python
d_df.info()
```

```python
d_df.dropna(subset="Engine Fuel Type",axis=0,inplace=True) 
```

```python
engine_mean = d_df['Engine HP'].mean()
eng_mode = d_df['Engine Cylinders'].mode()
mark_mode = d_df['Market Category'].mode()
```

```python
d_df['Engine HP'] = d_df['Engine HP'].fillna(engine_mean)
d_df['Engine Cylinders'] = d_df['Engine Cylinders'].fillna(eng_mode[0])
d_df['Market Category'] = d_df['Market Category'].fillna(mark_mode[0])
```

```python
d_df
```

```python
d_df.dropna(subset="Number of Doors",axis=0,inplace=True) 
```

```python
d_df.info()
```

```python
d_df.columns = d_df.columns.str.replace(' ', '_')
```

```python
d_df.columns
```

```python
d_df['Make'].unique()
```

```python
def clean_fuel(x):
    x = x.lower()
    if 'premium' in x or 'regular' in x:
        return 'petrol'
    elif 'diesel' in x:
        return 'diesel'
    elif 'electric' in x:
        return 'electric'
    elif 'flex' in x:
        return 'flex_fuel'
    elif 'natural gas' in x:
        return 'natural_gas'
    else:
        return 'other'

d_df['Fuel_Type'] = d_df['Engine_Fuel_Type'].apply(clean_fuel)
```

```python
d_df.drop('Engine_Fuel_Type', axis=1, inplace=True)
```

```python
def price_category(x):
    if x < 20000:
        return 'Low'
    elif x < 50000:
        return 'Medium'
    else:
        return 'High'

d_df['price_category'] = d_df['MSRP'].apply(price_category)
d_df.head()
```

```python
d_df['Driven_Wheels'].unique()
```

```python
d_df['Driven_Wheels'] = d_df['Driven_Wheels'].replace({'rear wheel drive' : 'RWD', 'front wheel drive' : 'FWD', 'all wheel drive' :'AWD',
 'four wheel drive':'4WD'})
```

```python
d_df.head()
```

```python
d_df['Transmission_Type'].unique()
```

```python
d_df['Transmission_Type'] = d_df['Transmission_Type'].replace({'AUTOMATED_MANUAL':'AUTOMATIC', 'DIRECT_DRIVE':'AUTOMATIC','UNKNOWN':'AUTOMATIC'})
```

```python
d_df['Market_Category'].unique()
```

```python
d_df['Market_Category'] = d_df['Market_Category'].str.split(',').str[-1]
```

```python
d_df['Vehicle_Style'].unique()
```

```python
d_df['Vehicle_Style'] = d_df['Vehicle_Style'].replace({'4dr Hatchback' : 'Hatchback',       '2dr Hatchback':'Hatchback',
             '4dr SUV': 'SUV',   'Passenger Minivan' :'Van',       'Cargo Minivan':'Van',
     'Crew Cab Pickup':'Pickup',  'Regular Cab Pickup':'Pickup', 'Extended Cab Pickup':'Pickup',
             '2dr SUV':'SUV',           'Cargo Van':'Van',     'Convertible SUV':'SUV',
       'Passenger Van':'Van'})
```

```python
d_df = d_df.reset_index(drop=True)
```

```python
d_df['avg_mpg'] = (d_df['city_mpg'] + d_df['highway_MPG']) / 2
```

```python
d_df.head(n=15)
```

```python
d_df.to_csv('cleaned_data/data.csv', index=False)
```
