## 1. Which top 10 car brands have the highest number of cars

## 2. Which brands have the highest average price (MSRP)?

```python
ord_MSRP=d_df.groupby(by='Make').agg({'MSRP' :'mean'}).sort_values(by='MSRP', ascending=False)
ord_MSRP.head(10)
```

**Observation**

The average sales maximum sales is Bugatti followed by Maybach and Rolls-Royce

## 3. For each brand, which model appears most frequently (proxy for most selling)?

```python
res = (d_df.groupby(['Make', 'Model']).size().reset_index(name='count')
          .sort_values(['Make', 'count'], ascending=[True, False]) )

top_model_per_brand = res.drop_duplicates('Make').sort_values(by='count', ascending=False)
top_model_per_brand.head(10)
```

**Observation**

The Chevrolet Brand silverado 1500 model appears most followed by Toyota Tundra and Ford F-150

## 4. How does price vary across vehicle styles (sedan, SUV, hatchback, etc.)?

```python
d_df.groupby('Vehicle_Style').agg({'MSRP':'mean'}).sort_values(by="MSRP",ascending=False).round(2)
```

**Observation**

The convertibles and coupes have the highest average prices, indicating their premium and performance-oriented nature. SUVs and sedans in the mid-price range, while hatchbacks, vans, and wagons are comparatively more affordable, reflecting their economy-focused positioning.
